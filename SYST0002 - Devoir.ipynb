```python
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

from numpy.linalg import eig
from numpy import arange, zeros, cos,ones
from scipy.signal import StateSpace, lsim, step, impulse, bode
from matplotlib.pyplot import plot,axis,show,xlabel,ylabel,title,grid,legend, semilogx

# data
d1 = 1/0.02
d2 = 1/0.002
k1 = d1
k2 = d2

# state space matrices
A = [[-d1,0.0],[k2,-d2]]
B = [[k1],[0.0]]
C = [0.0,1.0]
D = [0.0]

# linear time invariant system
syst = StateSpace(A,B,C,D)


if all(eig_value < 0 for eig_value in eig(A)[0]): print('Stable system')
else: print('Unstable system')
    

t = arange(0,0.10,0.001) 
u = zeros(len(t))

t1,y1,x1 = lsim(syst,u,t,[0.5,1])
t2,y2,x2 = lsim(syst,u,t,[2,1])
t3,y3,x3 = lsim(syst,u,t,[1,1])

plot(t,y1,'r--',label='(m,p)=(0.5,1)')
plot(t,y2,'b--',label='(m,p)=(2,1)')
plot(t,y3,'g--',label='(m,p)=(1,1)')

title('Simulated output acc. different initial conditions')
xlabel('time(s)')
ylabel('protein concentration')
legend()
grid(True)
show()
```

```python
t_step,y_step = step(syst,T=t)
plot(t,y_step)
title('Step response of the system')
xlabel('time(s)')
ylabel('amplitude')
grid(True)
show()

t_imp, y_imp = impulse(syst,T=t)
plot(t,y_imp)
title('Impulse response of the system')
xlabel('time(s)')
ylabel('amplitude')
grid(True)
show()
```

```python
t = arange(0,6,0.001)

t4,y4,x4 = lsim(syst,cos(t)    ,t)
t5,y5,x5 = lsim(syst,cos(10*t) ,t)
t6,y6,x6 = lsim(syst,cos(100*t),t)

plot(t,y4,'r',label='1 rad/s',linewidth=2.0)
plot(t,y5,'b',label='10 rad/s',linewidth=1.0)
plot(t,y6,'g',label='100 rad/s',linewidth=0.5)

title('Simulated output acc. different cosine inputs')
xlabel('time(s)')
ylabel('amplitude')
legend(bbox_to_anchor=(1.05, 1), loc=2)
show()
```

```python
print('Bode Diagram')

freq, amp, phase = bode(syst)
semilogx(freq,amp)
ylabel('Amplitude (dB)')
xlabel('Frequency')
grid(True)
show()

semilogx(freq,phase)
ylabel('Phase (deg)')
xlabel('Frequency')
grid(True)
show()
```
