```python

# 1.6 homework


%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

from numpy.linalg import *
from numpy import *
from scipy.signal import *
from matplotlib.pyplot import *

# data
d1 = 1/0.02
d2 = 1/0.002
k1 = d1
k2 = d2

# state space matrices
A = [[-d1,0.0],[k2,-d2]]
B = [[k1],[0.0]]
C = [0.0,1.0]
D = [0.0]

# linear time invariant system
syst = StateSpace(A,B,C,D)

# stability check
if all(eig_value < 0 for eig_value in eig(A)[0]): print('Stable system')
else: print('Unstable system') 
print('\n \n')    

# input and time scope definition
print('Free Response')
t = arange(0,0.10,0.001) 
u = zeros(len(t))

# simulated output 
t1,y1,x1 = lsim(syst,u,t,[0.5,1])
t2,y2,x2 = lsim(syst,u,t,[2,1])
t3,y3,x3 = lsim(syst,u,t,[1,1])

# graphs
plot(t,y1,'r',label='(m,p)=(0.5,1)')
plot(t,y2,'b',label='(m,p)=(2,1)')
plot(t,y3,'g',label='(m,p)=(1,1)')

title('Simulated output acc. different initial conditions on m')
xlabel('time(s)')
ylabel('protein concentration')
legend()
grid(True)
show()

# simulated output 
t1,y1,x1 = lsim(syst,u,t,[1,0.5])
t2,y2,x2 = lsim(syst,u,t,[1,2])
t3,y3,x3 = lsim(syst,u,t,[1,1])
t4,y4,x4 = lsim(syst,u,t,[1,0])

# graphs
plot(t,y1,'r',label='(m,p)=(1,0.5)')
plot(t,y2,'b',label='(m,p)=(1,2)')
plot(t,y3,'g',label='(m,p)=(1,1)')
plot(t,y4,'0.0',label='(m,p)=(1,0)')

title('Simulated output acc. different initial conditions on p')
xlabel('time(s)')
ylabel('protein concentration')
legend()
grid(True)
show()

# input and time scope definition
print('step Response')
t = arange(0,0.10,0.001) 
u = ones(len(t))

# simulated output 
t1,y1,x1 = lsim(syst,u,t,[0.5,1])
t2,y2,x2 = lsim(syst,u,t,[2,1])
t3,y3,x3 = lsim(syst,u,t,[1,1])

# graphs
plot(t,y1,'r',label='(m,p)=(0.5,1)')
plot(t,y2,'b',label='(m,p)=(2,1)')
plot(t,y3,'g',label='(m,p)=(1,1)')

title('Simulated output acc. different initial conditions on m')
xlabel('time(s)')
ylabel('protein concentration')
legend()
grid(True)
show()

# simulated output 
t1,y1,x1 = lsim(syst,u,t,[1,0.5])
t2,y2,x2 = lsim(syst,u,t,[1,2])
t3,y3,x3 = lsim(syst,u,t,[1,1])
t4,y4,x4 = lsim(syst,u,t,[1,0])

# graphs
plot(t,y1,'r',label='(m,p)=(1,0.5)')
plot(t,y2,'b',label='(m,p)=(1,2)')
plot(t,y3,'g',label='(m,p)=(1,1)')
plot(t,y4,'0.0',label='(m,p)=(1,0)')

title('Simulated output acc. different initial conditions')
xlabel('time(s)')
ylabel('protein concentration')
legend()
grid(True)
show()


```

```python

# homework 2.4


from scipy.integrate import odeint

def differential(x,t):
    d1 = 1/0.02
    k1 = d1
    d2 = 1/0.002
    k2 = d2
    K = 0.5
    n = 5

    dx = zeros(2)
    dx[0] = k1*(x[1]**n)/(K**n + x[1]**n)-d1*x[0]
    dx[1] = k2*x[0] - d2*x[1]
    return dx

# time scpoe
t = arange(0,0.10,0.001) 

# response of the system at each time t wrt to initial conditions in brackets
y21 = odeint(differential,[.5,1.],t)
y22 = odeint(differential,[2.,1.],t)
y23 = odeint(differential,[1.,1.],t)

# data
d1 = 1/0.02
d2 = 1/0.002
k1 = d1
k2 = d2

# state space matrices
A = [[-d1,0.0],[k2,-d2]]
B = [[k1],[0.0]]
C = [0.0,1.0]
D = [0.0]

# linear time invariant system
syst = StateSpace(A,B,C,D)

# input
u = ones(len(t))

# simulated output 
t1,y1,x1 = lsim(syst,u,t,[0.5,1])
t2,y2,x2 = lsim(syst,u,t,[2,1])
t3,y3,x3 = lsim(syst,u,t,[1,1])


# graphs
plot(t,y1,'r--',label='(m,p)=(0.5,1)')
plot(t,y2,'b--',label='(m,p)=(2,1)')
plot(t,y3,'g--',label='(m,p)=(1,1)')
plot(t,y21[:,1],'r',label=' (m,p)=(0.5,1), with auto-activation')
plot(t,y22[:,1],'b',label=' (m,p)=(2,1), with auto-activation')
plot(t,y23[:,1],'g',label=' (m,p)=(1,1), with auto-activation')

title('Simulated output acc. different initial conditions on m')
xlabel('time(s)')
ylabel('protein concentration')
legend(bbox_to_anchor=(1.05, 1), loc=2)
grid(True)
show()


# response of the system at each time t wrt to initial conditions in brackets
y21 = odeint(differential,[1.,.5],t)
y22 = odeint(differential,[1.,2.],t)
y23 = odeint(differential,[1.,1.],t)

# simulated output 
t1,y1,x1 = lsim(syst,u,t,[1,0.5])
t2,y2,x2 = lsim(syst,u,t,[1,2])
t3,y3,x3 = lsim(syst,u,t,[1,1])


# graphs
plot(t,y1,'r--',label='(m,p)=(1,0.5)')
plot(t,y2,'b--',label='(m,p)=(1,2)')
plot(t,y3,'g--',label='(m,p)=(1,1)')
plot(t,y21[:,1],'r',label='(m,p)=(1,0.5), with auto-activation')
plot(t,y22[:,1],'b',label='(m,p)=(1,2), with auto-activation')
plot(t,y23[:,1],'g',label='(m,p)=(1,1), with auto-activation')

title('Simulated output acc. different initial conditions on p')
xlabel('time(s)')
ylabel('protein concentration')
legend(bbox_to_anchor=(1.05, 1), loc=2)
grid(True)
show()
```
