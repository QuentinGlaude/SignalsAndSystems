```python
%matplotlib inline

import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

k1 = 1
k2 = 1
d1 = k1
d2 = k2

A = np.array([[-d1, 0], [-d2, k2]])
B = np.array([[k1], [0]])
C = np.array([[0, 1]])
D = np.array([[0]])

syst = signal.lti(A,B,C,D)
freq, amplitude, phase = signal.bode(syst)

plt.figure()
plt.semilogx(freq, amplitude)    # Bode magnitude plot
plt.ylabel('Amplitude (dB)')
plt.xlabel('Frequency')
plt.figure()
plt.semilogx(freq, phase)  # Bode phase plot
plt.ylabel('Phase (rad)')
plt.xlabel('Frequency')
plt.show()
```
